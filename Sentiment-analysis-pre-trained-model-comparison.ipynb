# 2. Run sentiment analysis predictions by using Pipeline

```python
# Using pipeline class to make predictions from models available in the Hub in an easy way
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
data = ["I love you", "I hate you", "the product would be great, had it not broke a minute after I purchased it"]
sentiment_pipeline(data)
```

```python
print(sentiment_pipeline.check_model_type,
sentiment_pipeline.framework,
sentiment_pipeline.model,
sentiment_pipeline.modelcard,
sentiment_pipeline.torch_dtype
)
```
